```python
from datetime import datetime
```

```python
datetime.fromtimestamp(int(df.iloc[0].timestampMs)/1000)
```

```python
arrive = datetime(2018,8,15)
```

```python
df.timestampMs = pd.to_numeric(df.timestampMs)
```

```python
def to_coord(df):
    return (df[['latitudeE7', 'longitudeE7']]/1e7).as_matrix()
```

```python
coords = to_coord(df[df.timestampMs > arrive.timestamp() * 1000])
```

```python
def heatmap_coords(coords):
    m = folium.Map(list(coords[0]), zoom_start=13)
    m.add_children(plugins.HeatMap(coords, radius=20))
    return m
```

```python
heatmap_coords(coords[:50000])
```
